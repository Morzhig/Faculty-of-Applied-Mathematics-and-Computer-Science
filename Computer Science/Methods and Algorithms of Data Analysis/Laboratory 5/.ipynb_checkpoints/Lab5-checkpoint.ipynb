## Задание 1. Детекция аномалий в данных кредитного скоринга.
Используйте набор данных кредитного скоринга с Kaggle ([ссылка на данные](https://www.kaggle.com/mlg-ulb/creditcardfraud)) и определите аномалии в данных, используя Z-score или IQR.

```python
df_credit = pd.read_csv(r'Datasets\creditcard.csv')

df_credit.head(7)
```

```python
df_credit.describe()
```

```python
def anomalies(df):
    df_anomalies = pd.DataFrame(columns=df.columns)
    
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1 

        df_anomalies[col] = df[(df[col] < Q1 - 3/2 * IQR) | (df[col] > Q3 + 3/2 * IQR)][col]

    return df_anomalies.reset_index().dropna(axis='columns', how='all')

df_anomalies = anomalies(df_credit.drop(columns=['Time','Class', 'Amount']))

df_anomalies.head(7)
```

## Задание 2. Поиск аномалий в данных кредитного скоринга.
Используйте набор данных кредитного скоринга с Kaggle ([ссылка на данные](https://www.kaggle.com/mlg-ulb/creditcardfraud)) и определите аномалии в данных, используя Z-score или IQR. 

```python
df_anomalies = pd.DataFrame(columns=df_credit.drop(columns=['Time','Class','Amount']).columns)

for col in df_anomalies.columns:
    df_anomalies[col] = df_credit[np.abs(scipy.stats.zscore(df_credit[col])) > 2][col]
            
df_anomalies.head(7)
```

## Задание 6. Нормализация и стандартизация признаков датасета "Red Wine Quality".
Для набора данных о качестве красного вина ([ссылка на данные](https://www.kaggle.com/uciml/red-wine-quality-cortez-et-al-2009)) примените различные методы нормализации и стандартизации к признакам, сравните результаты их применения.

```python
df_wine = pd.read_csv(r'Datasets\winequality-red.csv')

df_wine.head(7)
```

```python
df_wine.isna().sum()
```

```python
def minmax(df):
    for col in df.columns:
        df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

    return df

def mean_norm(df):
    for col in df.columns:
        df[col] = (df[col] - df[col].mean()) / (df[col].max() - df[col].min())

    return df

def z_standard(df):
    for col in df.columns:
        df[col] = (df[col] - df[col].mean()) / df[col].std()

    return df
```

```python
df_wine_norm_minmax = minmax(df_wine)

df_wine_norm_minmax.head(7)
```

```python
df_wine_norm_minmax.describe()
```

```python
df_wine_norm_mean = mean_norm(df_wine)

df_wine_norm_mean.head(7)
```

```python
df_wine_norm_mean.describe()
```

```python
df_wine_standard_z = z_standard(df_wine)

df_wine_standard_z.head(7)
```

```python
df_wine_standard_z.describe()
```

## Задание 8. Обработка текстовых данных в датасете "Twitter Airline Sentiment".
Набор данных "Twitter Airline Sentiment" на Kaggle ([ссылка на данные](https://www.kaggle.com/crowdflower/twitter-airline-sentiment)) содержит реальные твиты о различных авиакомпаниях. Примените техники предобработки текста, вроде удаления стоп-слов, стемминга, лемматизации и векторизации для подготовки данных к анализу.

```python
df_tweets = pd.read_csv(r'Datasets\Tweets.csv')

df_tweets.head(7)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    words = word_tokenize(text.lower())
    
    words = [word for word in words if word.isalnum() and word not in stop_words]
    
    words = [lemmatizer.lemmatize(stemmer.stem(word)) for word in words]
    
    return ' '.join(words)
```

```python
df_tweets['processed_text'] = df_tweets['text'].apply(preprocess_text)

df_tweets[['text', 'processed_text']].head(7)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

CountVectorizer = CountVectorizer()

X_countvectorized = CountVectorizer.fit_transform(df_tweets['processed_text'])

pd.DataFrame(X_countvectorized.toarray(), columns=CountVectorizer.get_feature_names_out()).head(7)
```
