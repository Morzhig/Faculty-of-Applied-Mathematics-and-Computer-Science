```python
import pandas as pd
import matplotlib.pyplot as plt
import pystan
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# Загрузка данных
ds = pd.read_excel(r'RETAIL.xlsx', decimal=',')
df = pd.DataFrame(ds, columns=['date', 'sales'])
df['date'] = pd.to_datetime(df['date'])

# Визуализация исходных данных
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['sales'])
plt.title('Продажи в сети бензоколонок в США')
plt.xlabel('Дата')
plt.ylabel('Продажи')
plt.show()

# Проверка на стационарность
result = adfuller(df['sales'])
print('ADF статистика:', result[0])
print('p-значение:', result[1])
print('Критические значения:')
for key, value in result[4].items():
    print(f'  {key}: {value}')

# Преобразование данных
df['non_trend_sales'] = df['sales'] - df['sales'].rolling(window=2).mean()
df['sales_diff'] = df['non_trend_sales'].diff()
df.dropna(inplace=True)

# Визуализация преобразованных данных
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['non_trend_sales'], label='Без тренда')
plt.plot(df['date'], df['sales_diff'], label='Стационарные')
plt.legend()
plt.title('Преобразованные продажи')
plt.xlabel('Дата')
plt.ylabel('Продажи')
plt.show()

# Проверка стационарности преобразованных данных
result = adfuller(df['sales_diff'])
print('ADF статистика:', result[0])
print('p-значение:', result[1])
print('Критические значения:')
for key, value in result[4].items():
    print(f'  {key}: {value}')

# Графики автокорреляции и частичной автокорреляции
plt.figure(figsize=(12, 6))
plt.subplot(211)
plot_acf(df['sales_diff'], lags=50, ax=plt.gca())
plt.title('Автокорреляция')

plt.subplot(212)
plot_pacf(df['sales_diff'], lags=50, ax=plt.gca())
plt.title('Частичная автокорреляция')

plt.tight_layout()
plt.show()

# Определение параметров модели ARIMA
p = 1
d = 1
q = 1

# Байесовская модель ARIMA с использованием PyStan
data = df['sales_diff'].values
stan_code = """
data {
  int<lower=0> T;         // Общая длина временного ряда
  real y[T];              // Временной ряд
}

parameters {
  real mu;                // Среднее значение
  real<lower=0> sigma;    // Стандартное отклонение ошибки
  real phi;               // Параметр AR
  real theta;             // Параметр MA
}

model {
  for (t in 2:T) {
    real y_hat = mu + phi * (y[t-1] - mu) + theta * (y[t-1] - mu);
    y[t] ~ normal(y_hat, sigma);
  }
}
"""
stan_data = {'T': len(data), 'y': data}

# Компиляция модели и подгонка данных
model = pystan.StanModel(model_code=stan_code)
fit = model.sampling(data=stan_data, iter=1000, chains=4)

# Визуализация результатов
fit.plot()
plt.show()

# Извлечение оценок параметров
params = fit.extract()
mu = params['mu'].mean()
sigma = params['sigma'].mean()
phi = params['phi'].mean()
theta = params['theta'].mean()

# Прогнозирование
n_forecast = 12
forecast = np.zeros(n_forecast)
y = data.tolist()

for t in range(len(y), len(y) + n_forecast):
    y_hat = mu + phi * (y[t-1] - mu) + theta * (y[t-1] - mu)
    forecast[t - len(y)] = y_hat
    y.append(y_hat)

# Рассчет MSE и MAE
mse = mean_squared_error(df['sales'][-12:], forecast)
mae = mean_absolute_error(df['sales'][-12:], forecast)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
```
