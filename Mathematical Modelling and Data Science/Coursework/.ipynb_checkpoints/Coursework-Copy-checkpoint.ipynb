```python
import pandas as pd
import matplotlib.pyplot as plt
import pystan
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# Ð—Ð°Ð³Ñ€ÑƒÐ·ÐºÐ° Ð´Ð°Ð½Ð½Ñ‹Ñ…
ds = pd.read_excel(r'RETAIL.xlsx', decimal=',')
df = pd.DataFrame(ds, columns=['date', 'sales'])
df['date'] = pd.to_datetime(df['date'])

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð¸ÑÑ…Ð¾Ð´Ð½Ñ‹Ñ… Ð´Ð°Ð½Ð½Ñ‹Ñ…
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['sales'])
plt.title('ÐŸÑ€Ð¾Ð´Ð°Ð¶Ð¸ Ð² ÑÐµÑ‚Ð¸ Ð±ÐµÐ½Ð·Ð¾ÐºÐ¾Ð»Ð¾Ð½Ð¾Ðº Ð² Ð¡Ð¨Ð')
plt.xlabel('Ð”Ð°Ñ‚Ð°')
plt.ylabel('ÐŸÑ€Ð¾Ð´Ð°Ð¶Ð¸')
plt.show()

# ÐŸÑ€Ð¾Ð²ÐµÑ€ÐºÐ° Ð½Ð° ÑÑ‚Ð°Ñ†Ð¸Ð¾Ð½Ð°Ñ€Ð½Ð¾ÑÑ‚ÑŒ
result = adfuller(df['sales'])
print('ADF ÑÑ‚Ð°Ñ‚Ð¸ÑÑ‚Ð¸ÐºÐ°:', result[0])
print('p-Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ:', result[1])
print('ÐšÑ€Ð¸Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¸Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ:')
for key, value in result[4].items():
    print(f'  {key}: {value}')

# ÐŸÑ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ð°Ð½Ð¸Ðµ Ð´Ð°Ð½Ð½Ñ‹Ñ…
df['non_trend_sales'] = df['sales'] - df['sales'].rolling(window=2).mean()
df['sales_diff'] = df['non_trend_sales'].diff()
df.dropna(inplace=True)

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð¿Ñ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ð°Ð½Ð½Ñ‹Ñ… Ð´Ð°Ð½Ð½Ñ‹Ñ…
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['non_trend_sales'], label='Ð‘ÐµÐ· Ñ‚Ñ€ÐµÐ½Ð´Ð°')
plt.plot(df['date'], df['sales_diff'], label='Ð¡Ñ‚Ð°Ñ†Ð¸Ð¾Ð½Ð°Ñ€Ð½Ñ‹Ðµ')
plt.legend()
plt.title('ÐŸÑ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ð°Ð½Ð½Ñ‹Ðµ Ð¿Ñ€Ð¾Ð´Ð°Ð¶Ð¸')
plt.xlabel('Ð”Ð°Ñ‚Ð°')
plt.ylabel('ÐŸÑ€Ð¾Ð´Ð°Ð¶Ð¸')
plt.show()

# ÐŸÑ€Ð¾Ð²ÐµÑ€ÐºÐ° ÑÑ‚Ð°Ñ†Ð¸Ð¾Ð½Ð°Ñ€Ð½Ð¾ÑÑ‚Ð¸ Ð¿Ñ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ð°Ð½Ð½Ñ‹Ñ… Ð´Ð°Ð½Ð½Ñ‹Ñ…
result = adfuller(df['sales_diff'])
print('ADF ÑÑ‚Ð°Ñ‚Ð¸ÑÑ‚Ð¸ÐºÐ°:', result[0])
print('p-Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ:', result[1])
print('ÐšÑ€Ð¸Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¸Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ:')
for key, value in result[4].items():
    print(f'  {key}: {value}')

# Ð“Ñ€Ð°Ñ„Ð¸ÐºÐ¸ Ð°Ð²Ñ‚Ð¾ÐºÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ð¸ Ð¸ Ñ‡Ð°ÑÑ‚Ð¸Ñ‡Ð½Ð¾Ð¹ Ð°Ð²Ñ‚Ð¾ÐºÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ð¸
plt.figure(figsize=(12, 6))
plt.subplot(211)
plot_acf(df['sales_diff'], lags=50, ax=plt.gca())
plt.title('ÐÐ²Ñ‚Ð¾ÐºÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ñ')

plt.subplot(212)
plot_pacf(df['sales_diff'], lags=50, ax=plt.gca())
plt.title('Ð§Ð°ÑÑ‚Ð¸Ñ‡Ð½Ð°Ñ Ð°Ð²Ñ‚Ð¾ÐºÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ñ')

plt.tight_layout()
plt.show()

# ÐžÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¾Ð² Ð¼Ð¾Ð´ÐµÐ»Ð¸ ARIMA
p = 1
d = 1
q = 1

# Ð‘Ð°Ð¹ÐµÑÐ¾Ð²ÑÐºÐ°Ñ Ð¼Ð¾Ð´ÐµÐ»ÑŒ ARIMA Ñ Ð¸ÑÐ¿Ð¾Ð»ÑŒÐ·Ð¾Ð²Ð°Ð½Ð¸ÐµÐ¼ PyStan
data = df['sales_diff'].values
stan_code = """
data {
  int<lower=0> T;         // ÐžÐ±Ñ‰Ð°Ñ Ð´Ð»Ð¸Ð½Ð° Ð²Ñ€ÐµÐ¼ÐµÐ½Ð½Ð¾Ð³Ð¾ Ñ€ÑÐ´Ð°
  real y[T];              // Ð’Ñ€ÐµÐ¼ÐµÐ½Ð½Ð¾Ð¹ Ñ€ÑÐ´
}

parameters {
  real mu;                // Ð¡Ñ€ÐµÐ´Ð½ÐµÐµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ
  real<lower=0> sigma;    // Ð¡Ñ‚Ð°Ð½Ð´Ð°Ñ€Ñ‚Ð½Ð¾Ðµ Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ðµ Ð¾ÑˆÐ¸Ð±ÐºÐ¸
  real phi;               // ÐŸÐ°Ñ€Ð°Ð¼ÐµÑ‚Ñ€ AR
  real theta;             // ÐŸÐ°Ñ€Ð°Ð¼ÐµÑ‚Ñ€ MA
}

model {
  for (t in 2:T) {
    real y_hat = mu + phi * (y[t-1] - mu) + theta * (y[t-1] - mu);
    y[t] ~ normal(y_hat, sigma);
  }
}
"""
stan_data = {'T': len(data), 'y': data}

# ÐšÐ¾Ð¼Ð¿Ð¸Ð»ÑÑ†Ð¸Ñ Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ð¸ Ð¿Ð¾Ð´Ð³Ð¾Ð½ÐºÐ° Ð´Ð°Ð½Ð½Ñ‹Ñ…
model = pystan.StanModel(model_code=stan_code)
fit = model.sampling(data=stan_data, iter=1000, chains=4)

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð¾Ð²
fit.plot()
plt.show()

# Ð˜Ð·Ð²Ð»ÐµÑ‡ÐµÐ½Ð¸Ðµ Ð¾Ñ†ÐµÐ½Ð¾Ðº Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¾Ð²
params = fit.extract()
mu = params['mu'].mean()
sigma = params['sigma'].mean()
phi = params['phi'].mean()
theta = params['theta'].mean()

# ÐŸÑ€Ð¾Ð³Ð½Ð¾Ð·Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð¸Ðµ
n_forecast = 12
forecast = np.zeros(n_forecast)
y = data.tolist()

for t in range(len(y), len(y) + n_forecast):
    y_hat = mu + phi * (y[t-1] - mu) + theta * (y[t-1] - mu)
    forecast[t - len(y)] = y_hat
    y.append(y_hat)

# Ð Ð°ÑÑÑ‡ÐµÑ‚ MSE Ð¸ MAE
mse = mean_squared_error(df['sales'][-12:], forecast)
mae = mean_absolute_error(df['sales'][-12:], forecast)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
```
