```python
import pandas as pd
import numpy as np
import pymc3 as pm

# Загрузка данных
data = pd.read_excel('dataset.xlsx')  # Замените 'data.csv' на путь к вашему файлу данных

# Создание копии данных с пропущенными значениями
missing_data = data.copy()

# Определение модели Байесовской многомерной регрессии
with pm.Model() as model:
    # Переменные
    x = pm.Normal('x', mu=0, sd=1, shape=data.shape[1]-1)
    mu = pm.Deterministic('mu', pm.math.dot(x, data.drop('X_4', axis=1).T))
    sigma = pm.HalfCauchy('sigma', beta=10)
    
    # Наблюдения
    sales_observed = pm.Normal('sales_observed', mu=mu, sd=sigma, observed=missing_data['X_4'].dropna())
    
    # Постериорное распределение
    trace = pm.sample(1000, tune=1000)
    
# Заполнение пропущенных значений
missing_indices = missing_data['X_4'].isnull()
missing_data.loc[missing_indices, 'X_4'] = pm.distributions.Deterministic('mu', pm.math.dot(trace['x'].mean(axis=0), missing_data.drop('X_4', axis=1).T)).random()

# Вывод результатов
print(missing_data)
```
