```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
```

Рассмотрим наш набор данных на предмет наличия пропусков и выберем тот столбец, в котором пропуск будет стоять на последнем месте.

```python
ds = pd.read_excel(r'dataset.xlsx', decimal=',')

df = pd.DataFrame(ds, columns=['X_1', 'X_2', 'X_3', 'X_4', 'X_5', 'X_6', 'X_7', 'X_8', 'X_9', 'Y'])

df
```

Как можем заметить, под наш критерий подходит столбец "X_4", рассмотрим его и столбец "Y".

```python
df.describe()
```

```python
df.mode(axis='rows', dropna=True, numeric_only=True)
```

```python
new_df = df

new_df
```

```python
n = new_df.isna().sum()

print(f'Количество пропусков в нашем наборе данных равно: {n}')
```

Перейдем к рассмотрению способов заполнения данных для пропусков, находящихся внутри набора данных, для пропуска, находящего на конце, попробуем спрогнозировать значение.

```python
new_df.median()
```

```python
new_df = new_df.fillna(new_df.mean())

new_df.to_excel(r'mydata.xlsx')

new_df
```

```python
new_df.describe()
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()

X = new_df['Y'][0:38].values.reshape(-1, 1)
y = new_df['X_4'][0:38]

clf = clf.fit(X, y)
```

```python
new_df['X_4'][38] = clf.predict(new_df['Y'][38].reshape(1,-1))

new_df[['X_4', 'Y']]
```

```python
plt.figure(figsize=(12,5))
plt.grid(True)
plt.autoscale(axis='x',tight=True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(y)
```
