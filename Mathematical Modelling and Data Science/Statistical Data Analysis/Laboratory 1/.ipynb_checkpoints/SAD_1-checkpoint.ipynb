```python
import pandas as pd
import scipy.stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import kstest

df = pd.read_excel(r'D:\Iris.xlsx')

sepalwid = df['sepalwid']

#Гистограмма без засорения
sns.distplot(sepalwid)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of sepalwid')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot(sepalwid)
Q1 = sepalwid.quantile(0.25)
Q3 = sepalwid.quantile(0.75)
IQR = Q3 - Q1
textstr = '\n'.join((
    r'$Median=%.2f$' % sepalwid.median(),
    r'$Q_1 =%.2f$' % Q1,
    r'$Q_3 =%.2f$' % Q3,
    r'$Q_1-1.5IQR =%.2f$' % (Q1 - 1.5 * IQR),
    r'$Q_3+1.5IQR =%.2f$' % (Q3 + 1.5 * IQR)))
ax.set_xlabel('Versicolor')
ax.set_ylabel('Sepal width')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
fig.tight_layout()
plt.show()
```

```python
fig = sm.qqplot(sepalwid, line='r')
plt.show()
```

```python
print(sepalwid.describe())
print(f"median: {sepalwid.median()}")
print(f"mode: {sepalwid.mode()}")
print(f"skewness: {sepalwid.skew()}")
print(f"kurtosis: {sepalwid.kurt()}")
```

```python
print(kstest(sepalwid, 'norm'))
print(scipy.stats.shapiro(sepalwid))
print(scipy.stats.chisquare(sepalwid))
```

```python
sepalwid = df['sepalwid']
outliers = df['petalwid']
contaminated_sample = sepalwid._append(pd.Series(outliers))
```

```python
sns.distplot(contaminated_sample)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of contaminated sepalwid')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot(contaminated_sample)
Q1 = contaminated_sample.quantile(0.25)
Q3 = contaminated_sample.quantile(0.75)
IQR = Q3 - Q1
textstr = '\n'.join((
     r'$Median=%.2f$' % contaminated_sample.median(),
     r'$Q_1 =%.2f$' % Q1,
     r'$Q_3 =%.2f$' % Q3,
     r'$Q_1-1.5IQR =%.2f$' % (Q1 - 1.5 * IQR),
     r'$Q_3+1.5IQR =%.2f$' % (Q3 + 1.5 * IQR)))
ax.set_xlabel('Versicolor')
ax.set_ylabel('Sepal width')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=14,
    verticalalignment='top', bbox=props)
fig.tight_layout()
plt.show()
```

```python
fig = sm.qqplot(contaminated_sample, line='r')
plt.show()
```

```python
print(contaminated_sample.describe())
print(f"median: {contaminated_sample.median()}")
print(f"mode: {contaminated_sample.mode()}")
print(f"skewness: {contaminated_sample.skew()}")
print(f"kurtosis: {contaminated_sample.kurt()}")
```

```python
print(kstest(contaminated_sample, 'norm'))
print(scipy.stats.shapiro(contaminated_sample))
print(scipy.stats.chisquare(contaminated_sample))
```
