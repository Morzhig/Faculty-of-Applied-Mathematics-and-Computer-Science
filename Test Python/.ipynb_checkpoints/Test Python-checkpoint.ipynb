```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
```

```python
url = 'https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json'

df = pd.read_json(url)

df
```

```python
df.describe()
```

```python
x1 = df['gt_corners']
x2 = df['rb_corners']

plt.figure(figsize=(20, 6))
plt.plot(x1, color='blue', label='gt_corners')
plt.plot(x2, color='red', label='rb_corners')
plt.title(f'График количества углов')
plt.xlabel('')
plt.ylabel('')
plt.legend(fontsize=15)
plt.show()
```

```python
mse = mean_squared_error(x2, x1)

print(mse)
```

```python
sns.pairplot(df, kind='reg')

plt.show()
```

```python
for column in df.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(df[column])
    plt.title(f'График для стобца {column}')
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['max'])
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
